```python
import zipfile
import os

# Define the path to the zip file and the destination directory
path_to_zip_file = 's3://ishaan-research/data/raw/adni/Multi-Visit_MRI_Cohort_Study_MRI.zip' # Replace with your file name/path
directory_to_extract_to = 's3://ishaan-research/data/raw/adni/Multi-Visit_MRI_Cohort_Study_MRI' # Replace with your target directory

# Ensure the destination directory exists (optional, extractall creates it if needed, but it's good practice for clarity)
# os.makedirs(directory_to_extract_to, exist_ok=True) 

# Open the zip file in read mode ('r') and extract all contents
with zipfile.ZipFile(path_to_zip_file, 'r') as zip_ref:
    zip_ref.extractall(directory_to_extract_to) 

print(f"Extracted all contents to {directory_to_extract_to}")
```
